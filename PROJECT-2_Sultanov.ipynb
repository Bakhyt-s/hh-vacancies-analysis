```python
# Задание 5.1: Работодатели на первом и пятом месте по количеству вакансий
query = '''
SELECT 
    e.name AS employer,
    COUNT(*) AS cnt
FROM public.vacancies v
JOIN public.employers e ON v.employer_id = e.id
GROUP BY e.name
ORDER BY cnt DESC
LIMIT 5
'''
df_5_1 = pd.read_sql_query(query, connection)
df_5_1
```

```python
# Задание 5.2: Количество работодателей и вакансий по каждому региону;
# среди регионов без вакансий — тот, где больше всего работодателей
query = '''
WITH emp_counts AS (
    SELECT area AS area_id, COUNT(*) AS employers_cnt
    FROM public.employers
    GROUP BY area
),
vac_counts AS (
    SELECT area_id, COUNT(*) AS vacancies_cnt
    FROM public.vacancies
    GROUP BY area_id
)
SELECT 
    a.name AS area,
    COALESCE(ec.employers_cnt, 0) AS employers_cnt,
    COALESCE(vc.vacancies_cnt, 0) AS vacancies_cnt
FROM public.areas a
LEFT JOIN emp_counts ec ON ec.area_id = a.id
LEFT JOIN vac_counts vc ON vc.area_id = a.id
WHERE COALESCE(vc.vacancies_cnt, 0) = 0
ORDER BY employers_cnt DESC
LIMIT 1
'''
df_5_2 = pd.read_sql_query(query, connection)
df_5_2
```

```python
# Задание 5.3: Количество уникальных регионов на работодателя, максимум
query = '''
SELECT 
    e.name AS employer,
    COUNT(DISTINCT v.area_id) AS regions_cnt
FROM public.vacancies v
JOIN public.employers e ON v.employer_id = e.id
GROUP BY e.name
ORDER BY regions_cnt DESC
LIMIT 1
'''
df_5_3 = pd.read_sql_query(query, connection)
df_5_3
```

```python
# Задание 5.4: Количество работодателей без указанной сферы деятельности
query = '''
SELECT COUNT(*) AS cnt
FROM public.employers e
LEFT JOIN public.employers_industries ei ON e.id = ei.employer_id
WHERE ei.employer_id IS NULL
'''
df_5_4 = pd.read_sql_query(query, connection)
df_5_4
```

```python
# Задание 5.5: Третья по алфавиту компания с ровно 4 сферами деятельности
query = '''
SELECT 
    e.name AS employer
FROM public.employers e
JOIN public.employers_industries ei ON e.id = ei.employer_id
GROUP BY e.name
HAVING COUNT(*) = 4
ORDER BY e.name
LIMIT 1 OFFSET 2
'''
df_5_5 = pd.read_sql_query(query, connection)
df_5_5
```

```python
# Задание 5.6: Количество работодателей со сферой "Разработка программного обеспечения"
query = '''
SELECT COUNT(*) AS cnt
FROM public.employers_industries ei
JOIN public.industries i ON ei.industry_id = i.id
WHERE i.name = 'Разработка программного обеспечения'
'''
df_5_6 = pd.read_sql_query(query, connection)
df_5_6
```

```python
# Задание 5.7: Вакансии Яндекса по регионам-миллионникам + строка Total
query = '''
WITH million_cities AS (
    SELECT unnest(ARRAY[
        'Москва','Санкт-Петербург','Новосибирск','Екатеринбург','Казань',
        'Нижний Новгород','Челябинск','Красноярск','Самара','Уфа',
        'Ростов-на-Дону','Омск','Краснодар','Воронеж','Пермь','Волгоград'
    ]) AS name
)
SELECT a.name, COUNT(*) AS cnt
FROM public.vacancies v
JOIN public.employers e ON v.employer_id = e.id
JOIN public.areas a ON v.area_id = a.id
WHERE e.name = 'Яндекс' AND a.name IN (SELECT name FROM million_cities)
GROUP BY a.name

UNION ALL

SELECT 'Total', COUNT(*)
FROM public.vacancies v
JOIN public.employers e ON v.employer_id = e.id
JOIN public.areas a ON v.area_id = a.id
WHERE e.name = 'Яндекс' AND a.name IN (SELECT name FROM million_cities)

ORDER BY cnt
'''
df_5_7 = pd.read_sql_query(query, connection)
df_5_7
```

Яндекс — безусловный лидер рынка вакансий: 1933 вакансии (в 4 раза больше второго места — Ростелекома с 491), присутствие в 181 регионе (тоже абсолютный максимум) и 485 вакансий только в городах-миллионниках. Это отражает масштаб компании как одного из крупнейших работодателей в IT-сфере региона с географически распределённой структурой найма.
Заметная часть базы работодателей (8419 из 23501, то есть почти 36%) вообще не указала сферу деятельности — это существенный пробел в данных, который стоит учитывать при любом отраслевом анализе: реальная картина по индустриям может отличаться от той, что видна по заполненным записям. Среди тех, кто сферу указал, «Разработка программного обеспечения» — одна из самых распространённых (3553 работодателя), что ожидаемо для базы, ориентированной на IT-вакансии.

```python
# Задание 6.1: Сколько вакансий имеет отношение к данным?
# Критерий: в названии есть 'data' или 'данн'
query = '''
SELECT COUNT(*) AS cnt
FROM public.vacancies
WHERE name ILIKE '%data%' OR name ILIKE '%данн%'
'''
df_6_1 = pd.read_sql_query(query, connection)
df_6_1
```

```python
# Задание 6.2: Вакансии для начинающего дата-сайентиста (уровень Junior)
query = '''
SELECT COUNT(*) AS cnt
FROM public.vacancies
WHERE (
    name ILIKE '%data scientist%'
    OR name ILIKE '%data science%'
    OR name ILIKE '%исследователь данных%'
    OR name ~* '\\yml\\y'
    OR name ILIKE '%machine learning%'
    OR name ILIKE '%машинн%обучен%'
)
AND (
    name ILIKE '%junior%'
    OR experience = 'Нет опыта'
    OR employment = 'Стажировка'
)
'''
df_6_2 = pd.read_sql_query(query, connection)
df_6_2
```

```python
# Задание 6.6: Средняя зарплата для DS по каждому типу опыта (без Junior)
query = '''
SELECT 
    experience,
    ROUND(AVG(
        (COALESCE(salary_from, salary_to) + COALESCE(salary_to, salary_from)) / 2
    )::numeric) AS avg_salary
FROM public.vacancies
WHERE (
    name ILIKE '%data scientist%'
    OR name ILIKE '%data science%'
    OR name ILIKE '%исследователь данных%'
    OR name ~* '\\yml\\y'
    OR name ILIKE '%machine learning%'
    OR name ILIKE '%машинн%обучен%'
)
AND NOT (
    name ILIKE '%junior%'
    OR experience = 'Нет опыта'
    OR employment = 'Стажировка'
)
AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)
GROUP BY experience
ORDER BY experience
'''
df_6_6 = pd.read_sql_query(query, connection)
df_6_6
```

Вакансий, так или иначе связанных с данными, довольно много — 1771, но специализированных именно под дата-сайентиста заметно меньше: лишь 51 из них ориентирована на начинающего специалиста (Junior/без опыта/стажировка), что говорит о том, что рынок DS-вакансий преимущественно рассчитан на специалистов уже с каким-то опытом, а не на новичков — вход в профессию через прямые junior-вакансии DS ограничен.
Зарплатный разрыв между уровнями опыта показателен: специалист с опытом от 1 до 3 лет может рассчитывать в среднем на 150 785 ₽, а с опытом от 3 до 6 лет — уже на 243 115 ₽, то есть рост почти в 1.6 раза. Это подчёркивает, что в DS практический опыт ценится особенно высоко — разница ощутимо больше, чем обычно наблюдается при переходе между аналогичными грейдами в других IT-специальностях.
Обращает на себя внимание отсутствие данных по зарплате для сегмента "Более 6 лет" среди DS-вакансий — либо такие вакансии крайне редко указывают зарплатную вилку в принципе, либо среди отфильтрованных DS-вакансий записей с таким опытом почти не осталось после исключения Junior-сегмента.

```python
# Задание 6.4: Количество DS-вакансий с ключевым навыком Python
query = '''
SELECT COUNT(*) AS cnt
FROM public.vacancies
WHERE (
    name ILIKE '%data scientist%'
    OR name ILIKE '%data science%'
    OR name ILIKE '%исследователь данных%'
    OR name ~* '\\yml\\y'
    OR name ILIKE '%machine learning%'
    OR name ILIKE '%машинн%обучен%'
)
AND NOT (
    name ILIKE '%junior%'
    OR experience = 'Нет опыта'
    OR employment = 'Стажировка'
)
AND key_skills ILIKE '%python%'
'''
df_6_4 = pd.read_sql_query(query, connection)
df_6_4
```

```python
# Задание 6.5: Среднее количество ключевых навыков в DS-вакансиях
query = '''
SELECT ROUND(AVG(array_length(string_to_array(key_skills, E'\\t'), 1))::numeric, 2) AS avg_skills
FROM public.vacancies
WHERE (
    name ILIKE '%data scientist%'
    OR name ILIKE '%data science%'
    OR name ILIKE '%исследователь данных%'
    OR name ~* '\\yml\\y'
    OR name ILIKE '%machine learning%'
    OR name ILIKE '%машинн%обучен%'
)
AND NOT (
    name ILIKE '%junior%'
    OR experience = 'Нет опыта'
    OR employment = 'Стажировка'
)
'''
df_6_5 = pd.read_sql_query(query, connection)
df_6_5
```

При выполнении заданий 6.3–6.5 обнаружено, что поле key_skills хранит навыки через символ табуляции (\t), а часть значений объединяет несколько технологий в одной строке (например, 'MS SQL/PostgreSQL', 'СУБД PostgreSQL', 'IPython', 'CPython'). Это создаёт неоднозначность при подсчёте: простой поиск подстроки (ILIKE '%SQL%', ILIKE '%python%') даёт результаты 168 и 308/310 соответственно, но такой подход рискует засчитывать смежные, но иные технологии (MySQL, IPython, CPython) как совпадения.
